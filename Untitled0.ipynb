```python
# Import libraries
import pandas
import pandas as pd
import matplotlib.pyplot as plt
# read the data from the CSV files into Pandas DataFrames
prodSalesDF = pandas.io.parsers.read_csv( 'product_sales_sample.csv' )
dayPartsDF = pandas.io.parsers.read_csv( 'dayparts_sample.csv' )
weatherDF = pandas.io.parsers.read_csv( 'weather_sample.csv' )
dmgrphcsDF = pandas.io.parsers.read_csv( 'demographics_sample.csv' )
```

```python
print prodSalesDF.columns
```

```python
# get the sales of each item
# group by the menu item and
# get total items sold (sum)
menuGroup = prodSalesDF.groupby('Menu Item')
itemsSold = menuGroup.sum()
# make Menu Item a column for ease of use rather than index
# Also clean up the menu item names by taking out numbers
# and other unwanted symbols like '-'
itemsSold['Menu Item'] = itemsSold.index\
        .map(lambda x: x.lstrip('0123456789-'))
# reset index to save space
itemsSold.reset_index(inplace=True, drop=True)
itemsSold = itemsSold[ [ 'Menu Item','POS Total Units Sold  Promo and Regular',\
       'POS Promotion Units Sold', 'POS Units Sold',\
       'POS Combo Units Sold' ] ]
# Get regular units sold : total - combo
itemsSold['POS Regular Units Sold'] = \
        itemsSold['POS Total Units Sold  Promo and Regular']\
    - itemsSold['POS Combo Units Sold']
# Get sum of sales of all items under different 
# categories like - total units sold, 
# regular units sold and combo units etc, this will
# help us to calculate sales as percentage rather than items
sumTotalItems = itemsSold['POS Total Units Sold  Promo and Regular'].sum()
sumPromoItems = itemsSold['POS Promotion Units Sold'].sum()
sumComboItems = itemsSold['POS Combo Units Sold'].sum()
sumRegItems = itemsSold['POS Regular Units Sold'].sum()
# get percentages and put them into diff set of cols
itemsSold['Perc Total Units Sold  Promo and Regular'] = \
    itemsSold['POS Total Units Sold  Promo and Regular']*100./sumTotalItems
itemsSold['Perc Promotion Units Sold'] = \
    itemsSold['POS Promotion Units Sold']*100./sumPromoItems
itemsSold['Perc Combo Units Sold'] = \
    itemsSold['POS Combo Units Sold']*100./sumComboItems
itemsSold['Perc Regular Units Sold'] = \
    itemsSold['POS Regular Units Sold']*100./sumRegItems
# sort according to total units sold promo and regular column
itemsSold.sort( \
               ['POS Total Units Sold  Promo and Regular'],\
               ascending=False, inplace=True )
# Plot
fig = plt.figure(figsize=(17,8))
ax1 = fig.add_subplot(111)
# plot total team wins
# ax1 = itemsSold.plot(x='Menu Item', y='POS Total Units Sold  Promo and Regular', kind='bar',\
#                  fontsize = 10, color='#1E90FF' )
# ax1 = itemsSold.plot(x='Menu Item', y='POS Promotion Units Sold', kind='bar',\
#                  fontsize = 10, color='green' )


itemsSold[['Perc Total Units Sold  Promo and Regular',\
           'Perc Regular Units Sold', 'Perc Combo Units Sold']].\
            plot( kind='bar', use_index=True,\
               ax=ax1, alpha=0.75,\
               grid=True, color=['aqua','orange','black'])


ax1.set_xlabel( 'Item Name', fontsize=14 )
ax1.set_ylabel( 'Units', fontsize=14 )
ax1.set_xticklabels(itemsSold['Menu Item'].values, fontsize=14)
ax1.set_title( 'Units sold for each menu item', fontsize=16 )
ax1.patch.set_facecolor('gray')
ax1.patch.set_alpha(0.75)
```
