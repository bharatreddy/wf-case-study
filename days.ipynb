```python
# Import libraries
import pandas
import matplotlib.pyplot as plt
import datetime
import numpy
import matplotlib.dates as mdates
# read the data from the CSV files into Pandas DataFrames
# prodSalesDF = pandas.io.parsers.read_csv( 'product_sales_sample.csv' )
dayPartsDF = pandas.io.parsers.read_csv( 'dayparts_sample.csv' )
# weatherDF = pandas.io.parsers.read_csv( 'weather_sample.csv' )
# dmgrphcsDF = pandas.io.parsers.read_csv( 'demographics_sample.csv' )
```

```python
print dayPartsDF.columns
# We'll work with datetime objects
# so convert Reporting Day to datetime obj
dayPartsDF['date'] = pandas.to_datetime( dayPartsDF['Reporting Day'] )
```

```python
# groupby by date to check for variations
dayGroup = dayPartsDF.groupby('date')
# get a sum of sales and transactions by days
daySales = dayGroup.sum()
# Divide data into two parts one in 2010 and the other
# in 2011 and then plot them. This makes it easy to
# visualize things.
date2011 = numpy.datetime64( datetime.date(2011,1,1) )
daySales2010 = daySales[ daySales.index < date2011 ]
daySales2011 = daySales[ daySales.index >= date2011 ]
```

```python
# Set some initial formatting for making plots
yearsLocPlot = mdates.YearLocator()   # every year
monthsLocPlot = mdates.MonthLocator()  # every month
daysLocPlot = mdates.DayLocator()
dateLocPlot = mdates.DateLocator()
dateTickFmt = mdates.DateFormatter('%d')
# Plot 2010 data
fig, (ax0, ax1) = plt.subplots(nrows=2, sharex=True, figsize=(11,8))
ax0.plot(daySales2010.index.tolist(), \
         daySales2010['Daypart Sales $'].tolist(),\
         'bo-')
ax0.set_ylabel('Sales ($)', fontsize=14)
ax0.set_title('Data for Jan-2010', fontsize=16)

ax1.plot(daySales2010.index.tolist(), \
         daySales2010['Daypart Transaction Qty'].tolist(),\
         'bo-')
ax1.set_ylabel('Transaction Qty', fontsize=14)

# format the ticks
ax0.xaxis.set_major_locator(daysLocPlot)
ax0.xaxis.set_major_formatter(dateTickFmt)
ax0.xaxis.set_minor_locator(daysLocPlot)

ax1.xaxis.set_major_locator(daysLocPlot)
ax1.xaxis.set_major_formatter(dateTickFmt)
ax1.xaxis.set_minor_locator(daysLocPlot)

datemin = daySales2010.index.min().to_datetime()
datemax = daySales2010.index.max().to_datetime()
ax1.set_xlim(datemin, datemax)

fig.savefig('figs/Jan2010-sales.png', bbox_inches='tight', \
                        pad_inches=0)
```

```python
# divide data into days 'Monday', 'Tuesday' etc
# and groupby them to see sales by weekdays.
daySales['weekday'] = daySales.index
daySales['weekdaynumber'] = daySales['weekday'].\
        apply( lambda x: x.weekday() )
daySales['weekday'] = daySales['weekday'].\
        apply( lambda x: x.to_datetime().strftime("%A") )
# Now groupby Weekday
dayNameGroup = daySales.groupby('weekday')
# get a sum of sales and transactions by dayparts
dayNameSales = dayNameGroup.sum()
```

```python
dayNameSales.sort( ['weekdaynumber'], inplace=True )
# Plot the data
fig, (ax0, ax1) = plt.subplots(nrows=2, sharex=True, figsize=(11,8))
ax0.bar(range(len(dayNameSales.index.tolist())), \
         dayNameSales['Daypart Sales $'].tolist(),\
         align='center', width=0.5,\
               color=['steelblue'])
# ax0.xticks(range(len(dayNameSales.index.tolist())), dayNameSales.index.tolist(), size='small')
ax0.set_ylabel('Sales ($)', fontsize=14)
ax0.grid(zorder=0)
ax0.set_title('Data by day of week', fontsize=16)

ax1.bar(range(len(dayNameSales.index.tolist())), \
         dayNameSales['Daypart Transaction Qty'].tolist(),\
         align='center', width=0.5,\
               color=['steelblue'])
ax1.set_ylabel('Transaction Qty', fontsize=14)
ax1.grid(zorder=0)
plt.xticks(range(len(dayNameSales.index.tolist())), dayNameSales.index.tolist(), fontsize=14)

fig.savefig('figs/sales-dayofweek.png', bbox_inches='tight', \
                        pad_inches=0)
```

```python
# check if the store is operating at
# different times of days for different days
# we can do this by grouping the data according
# to day of week and daypart
# Begin with getting weekday data for all dates
dayPartsDF['weekdaynumber'] = dayPartsDF['date'].\
        apply( lambda x: x.weekday() )
dayPartsDF['weekday'] = dayPartsDF['date'].\
        apply( lambda x: x.to_datetime().strftime("%A") )
# Groupby weekday and daypart name
dayNamePartGroup = dayPartsDF.groupby(['weekday', 'Daypart Name'])
dyNmKeyList = dayNamePartGroup.groups.keys()
# Make a scatter plot of the data
xAxis = [ i[0] for i in dayNamePartGroup.sum().index ]
yAxis = [ i[1] for i in dayNamePartGroup.sum().index ]

Xuniques, X = np.unique(xAxis, return_inverse=True)
Yuniques, Y = np.unique(yAxis, return_inverse=True)

salesDataDayPart = dayNamePartGroup.sum()['Daypart Sales $'].tolist()
xtnsDataDayPart = dayNamePartGroup.sum()['Daypart Transaction Qty'].tolist()
# get the transaction sizes as well
sizeIndsDayPart = []
for x,y in zip(xAxis, yAxis):
    sizeIndsDayPart.append( dayNamePartGroup.size().loc[ (x,y) ] )
```

```python
cm = plt.cm.get_cmap('RdYlBu')
fig = plt.figure(figsize=(11,8))
ax = fig.add_subplot(111)

sc = ax.scatter(X, Y, c=salesDataDayPart,cmap=cm,s=210)
ax.set(xticks=range(len(Xuniques)), xticklabels=Xuniques,
       yticks=range(len(Yuniques)), yticklabels=Yuniques) 
plt.colorbar(sc)
```
