```python
# Import libraries
import pandas
import matplotlib.pyplot as plt
import matplotlib as mpl
# read the data from the CSV files into Pandas DataFrames
prodSalesDF = pandas.io.parsers.read_csv( 'product_sales_sample.csv' )
# dayPartsDF = pandas.io.parsers.read_csv( 'dayparts_sample.csv' )
# weatherDF = pandas.io.parsers.read_csv( 'weather_sample.csv' )
# dmgrphcsDF = pandas.io.parsers.read_csv( 'demographics_sample.csv' )
```

```python
# Add some new columns
# Revenues : item_price * units_sold
prodSalesDF[ 'Revenue Total Units Sold  Promo and Regular' ] = \
        prodSalesDF[ 'POS Consumer Price' ] * prodSalesDF[ 'POS Total Units Sold  Promo and Regular' ]
prodSalesDF[ 'Revenue Promotion Units Sold' ] = \
        prodSalesDF[ 'POS Consumer Price' ] * prodSalesDF[ 'POS Promotion Units Sold' ]
prodSalesDF[ 'Revenue Combo Units Sold' ] = \
        prodSalesDF[ 'POS Consumer Price' ] * prodSalesDF[ 'POS Combo Units Sold' ]
prodSalesDF[ 'Revenue Units Sold' ] = \
        prodSalesDF[ 'POS Consumer Price' ] * prodSalesDF[ 'POS Units Sold' ]    
# Also clean up the menu item names by taking out numbers
# and other unwanted symbols like '-'
prodSalesDF['Menu Item'] = prodSalesDF['Menu Item']\
        .map(lambda x: x.lstrip('0123456789-'))
# Add some new columns
# We'll work with datetime objects
# so convert Reporting Day to datetime obj
prodSalesDF['date'] = pandas.to_datetime( prodSalesDF['Reporting Day'] )
# divide data into days 'Monday', 'Tuesday' etc
# and groupby them to see sales by weekdays.
prodSalesDF['weekday'] = prodSalesDF['date']
prodSalesDF['weekdaynumber'] = prodSalesDF['weekday'].\
        apply( lambda x: x.weekday() )
prodSalesDF['weekday'] = prodSalesDF['weekday'].\
        apply( lambda x: x.to_datetime().strftime("%A") )
# Sort by weekday number
prodSalesDF.sort( ['weekdaynumber'], inplace=True )
```
